#Numpy filter array

```python
import numpy as np
arr=np.array([41,42,43,44])
x=[True,False,True,False]
newarr=arr[x]
print(newarr)
```

##Creating the filter array

```python
#Create a filter array that will return only values higher than 42:
arr=np.array([41,42,43,44])
filter_arr=[]
for element in arr:
  if element>42:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
#create a filter array that will return only even elements from the original array:
arr=np.array([1,2,3,4,5,6,7])
filter_arr=[]
for element in arr:
  if element % 2 ==0:
    filter_arr.append(True)
  else:
    filter_arr.append(False)
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

##Creatin filter directly from array

```python
# create a filter array that will return only values higher than 42:
arr=np.array([41,42,43,44])
filter_arr=arr>42
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

```python
# create a filter array that will return only even elements fron the original array:
arr=np.array([1,2,3,4,5,6,7])
filter_arr=arr % 2==0
newarr=arr[filter_arr]
print(filter_arr)
print(newarr)
```

#InterQuartile Range(IQR)

```python
import numpy as np
import scipy.stats as stats

data=[12,15,14,10,18,20,22,25,14,17]

range_value=max(data)-min(data)
print(f"Range:{range_value}")
```

```python
variance_value=np.var(data,ddof=0)
sample_variance_value=np.var(data,ddof=1)
print(f"population variance:{variance_value}")
print(f"sample variance:{sample_variance_value}")
```

```python
std_dev_value=np.std(data,ddof=0)
sample_std_dev_value=np.std(data,ddof=1)
print(f"population standard deviation:{std_dev_value}")
print(f"sample standard deviation:{sample_std_dev_value}")
```

```python
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
iqr_value=Q3-Q1
print(f"Interquartile Range(IQR):{iqr_value}")
```
